### Pytorch - classifier
- CNN을 활용한 모델을 만들어서 CIFAR-10에 대하여 정확도 95% 이상의 Classifier를 만드세요
- 예제 코드는 MNIST에 대한 코드

```python
from matplotlib import pyplot as plt
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor
import torch.nn as nn
from torch import optim
from torch.utils.data import DataLoader

torch.manual_seed(0) # seed 고정 - 여러번 실행해도 같은 결과가 나오도록 설정
```

#### Dataset download

```python
train_data = datasets.CIFAR10(
    root = 'data',
    train = True,                         
    transform = ToTensor(), 
    download = True,            
)
dataloader_train = DataLoader(train_data, 
                              batch_size=32, 
                              shuffle=True)

test_data = datasets.CIFAR10(
    root = 'data', 
    train = False, 
    transform = ToTensor()
)
dataloader_test = DataLoader(test_data, 
                              batch_size=32, 
                              shuffle=True)

# CIFAR10
# train_data = datasets.CIFAR10(
#     root = 'data',
#     train = True,                         
#     transform = ToTensor(), 
#     download = True,            
# )
# dataloader_train = DataLoader(train_data, 
#                               batch_size=16, 
#                               shuffle=True)
```

#### Check images

```python
figure = plt.figure(figsize=(10, 8))
cols, rows = 5, 5
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(len(train_data), size=(1,)).item()
    img, label = train_data[sample_idx]
    figure.add_subplot(rows, cols, i)
    plt.title(label)
    plt.axis("off")
    plt.imshow(img.squeeze().T, cmap="gray")
plt.show()
print('Data shape:', img.shape)
```

#### CNN

```python
class CNN(nn.Module):
    def __init__(self):
        ## TODO ##
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels = 3,
                              out_channels = 32,
                              kernel_size = 3,
                              stride = 1,
                              padding = 1)
        self.relu = nn.ReLU()
        self.maxpool = nn.MaxPool2d(kernel_size = 2,
                                   stride = 2)
        self.conv2 = nn.Conv2d(in_channels = 32,
                              out_channels = 32,
                              kernel_size = 5,
                              stride = 1,
                              padding = 2)
        
        self.conv3 = nn.Conv2d(in_channels = 32,
                              out_channels = 64,
                              kernel_size = 3,
                              stride = 1,
                              padding = 1)
        
        self.conv4 = nn.Conv2d(in_channels = 64,
                              out_channels = 64,
                              kernel_size = 5,
                              stride = 1,
                              padding = 2)
        
        self.dropout25 = nn.Dropout2d(0.25)
        self.dropout5 = nn.Dropout2d(0.5)
        
#         self.linear = nn.Linear(in_features = 64*8*8,
#                                out_features = 512)
        self.out = nn.Linear(in_features = 64*8*8,
                               out_features = 10)
    def forward(self, x):
        ## TODO ##
        
        x = self.conv1(x)
        x = self.relu(x)
        x = self.conv2(x)
        x = self.relu(x)
        x = self.maxpool(x)
        x = self.dropout25(x)

        x = self.conv3(x)
        x = self.relu(x)
        x = self.conv4(x)
        x = self.relu(x)
        x = self.maxpool(x)
        x = self.dropout25(x)

        x = x.view(x.shape[0], -1)
#         x = self.linear(x)
#         x = self.relu(x)
#         x = self.dropout5(x)
        x = self.out(x)

        return x

x = torch.randn(16,3,32,32)
cnn = CNN()

cnn(x)
```

#### Train

```python
cnn = CNN()
loss_func = nn.CrossEntropyLoss()   
optimizer = optim.Adam(cnn.parameters(), lr = 0.001)   
```

```python
# Train the model
num_epochs = 5

for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(dataloader_train):
        output = cnn(images)           
        loss = loss_func(output, labels)

        # clear gradients for this training step   
        optimizer.zero_grad()           

        # backpropagation, compute gradients 
        loss.backward()    
        # apply gradients             
        optimizer.step()                

        if (i+1) % 100 == 0:
            print ('Epoch [{}/{}], Step [{}], Loss: {:.4f}' 
                   .format(epoch + 1, num_epochs, i + 1, loss.item()))
```

#### Test

```python
cnn.eval()
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in dataloader_test:
        test_output = cnn(images)
        pred_y = torch.max(test_output, 1)[1].data.squeeze() # max 를 통해 가장 확률이 높은 숫자를 선택
        correct += (pred_y == labels).sum().item() # 가장 확률이 높은 숫자인 pred_y와 label이 같으면 정답 - 정답인 개수 확인
        total += labels.size(0) 
    accuracy = correct / total * 100
    
print('Test Accuracy of the model on the 10000 test images: %.2f Percent' % accuracy)
```
