```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn import metrics
```

### 실습 과제 1. 데이터 불러온 후 NaN 값이 있다면 해당 행 제거하고, train_data : test_data = 7:3 split
#### 종속변수는 PKT_Y, 나머지 변수들이 독립변수

```python
df = pd.read_csv('contest.csv')
df.info()
rdf = df.drop(columns = ['No.','LOT_ID','WF_ID','X_AXIS','Y_AXIS','GRADE_CD'], axis = 1)
```

```python
rdf = rdf.dropna(subset=['PKT_Y'], how = 'any')
rdf.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(rdf.drop('PKT_Y',axis = 1),rdf.loc[:,'PKT_Y'], test_size = 0.3, random_state = 10)
```

### 실습 과제 2. 선형 커널, 3차항 커널, RBF 커널, 시그모이드 커널로 각각 SVM 모델 구현 후 accuracy, F1 score, confusion matrix, precision, recall (C=1, gamma = 0.1)

```python
from sklearn import svm
kernels = ['linear','poly','rbf','sigmoid']

for k in kernels:
    svm_model = svm.SVC(kernel = k, C = 1, gamma = 0.1)
    svm_model.fit(X_train,y_train)
    y_pred = svm_model.predict(X_test)
    
    con_mat = metrics.confusion_matrix(y_test,y_pred)
    acc = metrics.accuracy_score(y_test,y_pred)
    f1 = metrics.f1_score(y_test,y_pred)
    
    print("Kernel : ", k)
    print("confusion_matrix : ", con_mat)
    print("accuracy : ", acc)
    print("f1_score : ", f1)
    print("="*30)
```

### 실습 과제 3. 하이퍼 파라미터 C 값(0.01, 0.1, 1, 10)에 따른 RBF 커널을 SVM 모델의 accuracy와 F1 score 변화 관찰

```python
C_list = [0.01,0.1,1,10,100]

for C_value in C_list:
    svm_model = svm.SVC(kernel = 'rbf', C = C_value, gamma = 0.1)
    svm_model.fit(X_train,y_train)
    y_pred = svm_model.predict(X_test)
    
    con_mat = metrics.confusion_matrix(y_test,y_pred)
    acc = metrics.accuracy_score(y_test,y_pred)
    f1 = metrics.f1_score(y_test,y_pred)
    
    print('C value : ',C_value)
    #print("confusion_matrix : ", con_mat)
    print("accuracy : ", acc)
    print("f1_score : ", f1)
    print("="*30)
```

### 실습 과제 4. 하이퍼 파라미터 gamma 값(0.005,0.01,0.1, 1)에 따른 RBF 커널을 SVM 모델의 accuracy와 F1 score 변화 관찰

```python
gamma_list = [0.01,0.1,1,10,100]

for gam in gamma_list:
    svm_model = svm.SVC(kernel = 'rbf', C = 1, gamma = gam)
    svm_model.fit(X_train,y_train)
    y_pred = svm_model.predict(X_test)
    
    con_mat = metrics.confusion_matrix(y_test,y_pred)
    acc = metrics.accuracy_score(y_test,y_pred)
    f1 = metrics.f1_score(y_test,y_pred)
    
    print('gamma value : ',gam)
    #print("confusion_matrix : ", con_mat)
    print("accuracy : ", acc)
    print("f1_score : ", f1)
    print("="*30)
```

### 실습 과제 5. 하이퍼 파라미터 degree(1,2,3,4,5)에 따른 다항식 커널 SVM 모델 accuracy와 F1 score 변화 관찰

```python
degree_list = [1,2,3,4,5]

for deg in degree_list:
    svm_model = svm.SVC(kernel = 'poly',degree = deg, C = 1, gamma = 0.1)
    svm_model.fit(X_train,y_train)
    y_pred = svm_model.predict(X_test)
    
    con_mat = metrics.confusion_matrix(y_test,y_pred)
    acc = metrics.accuracy_score(y_test,y_pred)
    f1 = metrics.f1_score(y_test,y_pred)
    
    print('degree : ',deg)
    #print("confusion_matrix : ", con_mat)
    print("accuracy : ", acc)
    print("f1_score : ", f1)
    print("="*30)
```
