```python
#5. your code here

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty = 'l2')
lr.fit(train_img,train_lbl)
```

Model Performance Evaluation

```python
#6. Accuracy를 계산하라.
```

```python
#6. your code here

lbl_train_pred = lr.predict(train_img)
lbl_test_pred = lr.predict(test_img)

print("Train acc : ", accuracy_score(train_lbl,lbl_train_pred))
print("Test acc : ", accuracy_score(test_lbl,lbl_test_pred))
```
